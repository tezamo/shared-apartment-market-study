## 7. Regression analysis

````python
import statsmodels.api as sm
import plotly.express as px

# Define the explained and explanatory variables
X = wg_data_df[['Size', 'WG Type', 'Online Duration']]
y = wg_data_df['Price']

# Add a constant 
X = sm.add_constant(X)

# Perform the OLS 
model = sm.OLS(y, X).fit()

# Print the OLS  
print(model.summary())
````

#### Interpretation : 
Intercept (**801**): When Size and WG Type are zero, the average Price is approximately **801** euros.  

Size (**1.07**): For each additional square meter, the Price increases by approximately **1.07** euros, which is statistically significant.  

WG Type (**-65**): For each additional room in the WG (WG Type), the Price decreases by approximately **65** euros, which is statistically significant. 

Online duration (**-2**): Each additional minute , decrease the price about **2** euros.

Overall Model: The model explains about **6%** of the variance in prices.

````python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Load the data from the Excel file
df = pd.read_excel('/workspaces/fdap-2024-mortezamotegh/apartment_analysis/wg_listings_berlin_modified.xlsx')

# Select the dependent variable (Online Duration) and independent variables
X = df[['Size', 'WG Type', 'Online Duration']]
y = df['Price']

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Fit an OLS regression model
model = sm.OLS(y, X).fit()

# Predict the durations using the fitted model
predictions = model.predict(X)

# Create a scatter plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(y, predictions, color='blue', alpha=0.5, label='Predicted vs Actual')  # Actual vs Predicted scatter plot
plt.plot(y, y, color='red', linestyle='--', linewidth=2, label='Perfect Fit')  # Line for perfect fit

# Adding labels and title
plt.title('Predicted vs Actual Online Durations')
plt.xlabel('Actual Online Duration (minutes)')
plt.ylabel('Predicted Online Duration (minutes)')
plt.legend(title='Legend')
plt.grid(True)

# Show the plot
plt.show()
````
